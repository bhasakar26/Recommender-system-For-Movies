```python
movies_df=pd.read_csv(r'C:\Users\Bhaskar\Desktop\files\Movie-Analysis-master\data\tmdb_5000_movies.csv')
```

```python
credit_column_renamed=credits.rename(index=str,columns={'movie_id':'id'})
movies_df_merge=movies_df.merge(credit_column_renamed,on='id')
```

```python
movies_df_merge.head()
```

```python
movies_df_merge.shape
```

```python
movies_df_merge=movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
```

```python
movies_df_merge.columns
```

```python
movies_df_merge.head()
```

```python
movies_df_merge.info()
```

```python
movies_df_merge.isnull().sum()
```

```python
V=movies_df_merge['vote_count']
R=movies_df_merge['vote_average']
C=movies_df_merge['vote_average'].mean()
M=movies_df_merge['vote_count'].quantile(0.70)
```

```python
movies_df_merge['weight_average']=((R*V)+(C*M))/(V+M)
```

```python
movies_df_merge.head()
```

```python
movie_sorted_ranking=movies_df_merge.sort_values('weight_average',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weight_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weight_average=movies_df_merge.sort_values('weight_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weight_average'].head(10),y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.savefig('best_movies_png')
```

```python
popularity=movies_df_merge.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity)

plt.savefig('best_movies_png')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaing=MinMaxScaler()
movie_scaling_ranking=scaing.fit_transform(movies_df_merge[['weight_average','popularity']])
movie_normalised=pd.DataFrame(movie_scaling_ranking,columns=[['weight_average','popularity']])
```

```python
movie_normalised.head()
```

```python
movies_df_merge[['normalised_weight_average','normalised_popularity']]=movie_normalised
```

```python
movies_df_merge.head()
```

```python
movies_df_merge['score']=movies_df_merge['normalised_weight_average']*0.5+movies_df_merge['normalised_popularity']*0.5
```

```python
movies_df_merge=movies_df_merge.sort_values(['score'],ascending=False)
```

```python
movies_df_merge[['original_title','normalised_weight_average','normalised_popularity','score']].head(20)
```

```python
score_df=movies_df_merge.sort_values(['score'],ascending=False)
plt.figure(figsize=(16,7))
ax=sns.barplot(x=score_df['score'].head(10),y=score_df['original_title'].head(10),data=score_df)
```
